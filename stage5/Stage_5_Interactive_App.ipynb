## 5. Using your real Spotify data

To use this app on **your own** listening history (not just sample data):

1. Go to https://www.spotify.com/account/privacy/
2. Scroll to "Download your data" → click "Request"
3. Wait 1–30 days for Spotify to email you a ZIP file
4. Unzip it — find `StreamingHistory_music_*.json` files
5. Copy them into this folder
6. Pass the file name to `run()`:
